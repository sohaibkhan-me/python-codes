```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# line plot

x = np.linspace(0,10,5)
y = np.cos(x)

plt.xlabel(" input ")
plt.ylabel("cos function output ")

plt.title("line plot cos function vs input ")
plt.plot(x,y)
plt.show()
```

```python
# scatter plot

x = np.linspace(0,10,5)
y = np.cos(x)

plt.xlabel(" input ")
plt.ylabel("cos function output ")

plt.title("line plot cos function vs input ")

plt.scatter(x,y)

plt.show()
```

```python
x = np.array([2,4,6,8])
y = np.array([2,4,6,8])

plt.xlabel(" input ")

plt.ylabel(" output ")

plt.title("bar plot ")

plt.bar(x,y,color = "green")

plt.show()

```

```python
x = np.array([2,4,6,8,10])

plt.hist(x,bins = 3)
plt.show()
```

```python
labels = ["OOP", "Data structure", "Algorathim", "Artificial intelligence"]
sizes = [15, 30, 45, 10]

plt.pie(sizes,labels=labels, autopct='%1.1f%%')
plt.title("Total students in per subject Pie Chart")
plt.show()
```
