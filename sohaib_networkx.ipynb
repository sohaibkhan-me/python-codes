```python
import networkx as nx
```

```python
graph = nx.Graph()
```

```python
graph.add_node(1)

graph.add_node(2)

graph.add_node(3)

graph.add_node(4)

graph.add_edge(1,2)
graph.add_edge(2,3)
graph.add_edge(3,4)
graph.add_edge(4,1)

nx.draw(graph, with_labels=True, node_color="blue", node_size=2000)
```

```python
graph = nx.Graph()

graph.add_nodes_from([1,2,3,4,5])

graph.add_edges_from([(1,2),(3,4),(4,5),(2,3)])

nx.draw(graph, with_labels=True, node_color="Red", node_size=2000)
```

```python

import matplotlib.pyplot as plt

graph = nx.Graph()
graph.add_nodes_from(["peshawar", "mardan", "charsada", "lahore", "karachi"])

graph.add_weighted_edges_from([
    ("peshawar", "mardan", 60),
    ("peshawar", "charsada", 40),
    ("peshawar", "lahore", 470),
    ("lahore", "karachi", 1210),
    ("mardan", "charsada", 35),
])

pos = nx.spring_layout(graph)  # layout positions for nodes
nx.draw(graph, pos, with_labels=True, node_color="red", node_size=2800)

edge_labels = nx.get_edge_attributes(graph, "weight")
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)

shortest_path = nx.shortest_path(graph, source="peshawar", target="karachi", weight="weight")
shortest_distance = nx.shortest_path_length(graph, source="peshawar", target="karachi", weight="weight")

print("Shortest Path:", shortest_path)
print("Distance:", shortest_distance)

plt.show()
```
